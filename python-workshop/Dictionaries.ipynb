# Dictionary

A data structure that stores elements as a “key-value” pair. 



### Creating Dictionary
Creating a dictionary is as simple as placing items inside curly braces {} separated by commas.

```python
# 3 ways to create
# Method 1 - create empty dictionary
my_dictionary = dict()
new_dictionary = {}

# Method 2 - create dictionary with specific key value pair {key1:value2 , key2:value2}
programming_languages = {1 : "C", 2: "C++", 3: "Python"}

#Method 3 - create using dict() method
cities = dict([(1, "Mangalore"), (2, "Bangalore")]) 

print(my_dictionary)
print(new_dictionary)
print(programming_languages)
print(cities)
```

### Access Dictionary Elements

- Keys can be used either inside square brackets [] or with the get() method.
- If we use the square brackets **[ ]**, **KeyError** is raised in case a **key is not found in the dictionary**. On the other hand, the **get()** method returns **None if the key is not found**.

```python
# get vs [] for retrieving elements
record = {'name': 'Python', 'age': 29}

# Output: Python
print(record['age'])

# Output: 29
print(record.get('age',30))

# Trying to access keys which doesn't exist throws error
# Output None
print(record.get('address'))

# KeyError
print(record['address'])
```

#### Nested Dictionary
A dictionary within a dictionary is called as nested dictionary


```python
movies = {
    "Hollywood": {
        "Marvel" : ["Thor","Ant Man", "Iron Man"]
    },
    "Bollywood": ["Uri","Badla","Raazi"]
}

#To access Thor -> Chain the access brackets till you reach correct ones
print(movies["Hollywood"]["Marvel"][2])

#To access Raazi
print(movies["Bollywood"][2])
```

### Remove Elements from Dictionary

```python
movies = { "Hollywood" : {"Marvel" : ["Thor", "Ant Man", "Iron Man"]}, "Bollywood" : ["Uri", "Badla", "Raazi"]}
# Method 1 -- Using del
del movies["Bollywood"]
del movies["Hollywood"]["Marvel"]
print(movies)                                                                                      
                                                                                      
# Method 2 -- Using pop(index)
programming_languages = {1 : "C", 2: "C++", 3: "Python"}
programming_languages.pop(2) # Removes item based on index --> index 2 means 2nd element 
                                                                                      
print(programming_languages)
                                                                                      
# Method 3 -- Using popitem to remove last key value pair
programming_languages.popitem() # Removes last element
print(programming_languages)
                                                                                      
# Method 4-- Clears all elements
programming_languages.clear()
print(programming_languages)
```

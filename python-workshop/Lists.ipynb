### Mixed List, Get List Length and Print Elements

```python
# Mixed List with numbers, strings, boolean and decimals
a_list = [1, "Some string here", True, 3.1416]
print(a_list)
print("The contents of the list are:\t", a_list)    #\t is for giving tab space

print("***************")
#Get list Length
print("Length of a_list", len(a_list))
print("***************")

# Print elements one by one
for element in a_list:
    print(element)
print("***************")

# Print type of elements one by one
for element in a_list:
    print(type(element))
print("***************")
```

### List Indexing - Access Elements

#### Normal

```python
animals = ["penguin", "cow", "camel"]

print( animals[0] )
print( animals[1] )
print( animals[2] )
```

Lists start counting at **zero** in python from left to right. First element is in 0th position, 2nd in 1st position and so on


```python
for index in range(0, len(animals)):
    print("'"+str(animals[index]) +"' is at position "+ str(index) +" in list")
```

#### Negative Accessing (to access from right to left)
Python allows negative indexing for its list. The index of -1 refers to the last item, -2 to the second last item and so on.


```python
# Negative indexing in lists
my_list = ['p','r','o','b','e']

print(my_list[-3])
```

### Inserting into a list

```python
# Lets consider a new list of super villains now
super_villains = ["The Joker", "Doctor Doom", "Loki", "Sinestro"]

#To add at the end of list -> use append(value)"
super_villains.append("Thanos")   
print("After appending")
print(super_villains)

#To insert at 3rd position--> use insert(position, value)
super_villains.insert(2, "Dr. Octopus")   # Remember list starts from 0
print("After Inserting")
print(super_villains)
```

#### Remove Elements from List

```python
# Lets consider Super Villains example
# Pop removes the last element
super_villains.pop()
print(super_villains)

# Remove removes the specific element
super_villains.remove("Doctor Doom")
print(super_villains)
```

#### List Slicing

```python
hello_list = ['h','e','0','l','o']
sliced_list = hello_list[-2:-1]
print(sliced_list)
```

hello_list[1:4] - The 1 means to start at second element in the list (Since index starts at 0). The 4 means to end at the fifth element in the list, but not include it. 

#### Slicing extracts some elements, based on a start and stop.  ':' operator is called the slicing operator.


#### Reverse a list

```python
vowels= ['a','e','i','o','u']
#.reverse() will just reverse the elements of the list
vowels.reverse()
print(vowels)
```

## Applications

#### Find sum of all elements

```python
# Find sum of all elements in list
cost = [0.424, 0.4221, 0.4185, 0.4132, 0.413]

def find_sum(cost):
    # Method 1
    total_inbuilt = sum(cost)
    print(total_inbuilt)

    # Method 2 - Use normal for loop
    sum_total = 0
    for element in cost:
        sum_total = sum_total + element #0+0.424 #0.424+0.4221
    print(sum_total)
    
find_sum(cost)
```

#### Given a list of alphabets check whether alphabet J exists in the list.


```python
import string


# Search for an element in list
def linear_search(list, key):
    for element in list:
        if element == key:
            # If key matches
            return True
    # If key is not present at the end of list
    return False

#get string of upper case alphabets in string
alphabet_string = string.ascii_uppercase
alphabet_list=[] 
# Convert to list
alphabet_list[:0]=alphabet_string
print(alphabet_list)
linear_search(alphabet_list,'K')
```
